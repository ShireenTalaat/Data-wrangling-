```python
# Create the Twitter API object
consumer_key = 'HIDEN'
consumer_secret = 'HIDEN'
access_token = 'HIDEN'
access_secret = 'HIDEN'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
# As explained in project walkthrough 
exp_tweet = api.get_status(twitter_archive_df.tweet_id[100], tweet_mode = 'extended')
content = exp_tweet._json
content
```

```python
# As explianed in Data hub (Q & A)

# Creating a file for the tweets' text data
errors = []
if not os.path.isfile('tweet_json.txt'):
    # create the file and write on it
    with open ('tweet_json.txt', 'w') as file:
        for tweet_id in twitter_archive_df['tweet_id']:
            try:
                status = api.get_status(tweet_id, wait_on_rate_limit=True,  wait_on_rate_limit_notify=True, tweet_mode = 'extended')
                json.dump(status._json, file)
                file.write('\n')
            except Exception as e:
                print("Error on tweet id {}".format(tweet_id) + ";" + str(e))
                errors.append(tweet_id)
```

```python
# stack over flow example

#query all of the tweet IDs in the WeRateDogs Twitter archive
tweet_ids = twitter_archive_df.tweet_id.values
print("# of tweet_ids: " + str(len(tweet_ids)) + "\n")

count = 0
tweet_json_list = []
fails_dict = {}

for tweet_id in tweet_ids:
    count += 1
    try:
        # get tweet's JSON data then append to the tweet JSON list
        tweet = api.get_status(tweet_id, tweet_mode = 'extended')
        tweet_json_list.append(tweet._json)
    except tweepy.TweepError as error:
        # save the error to the fail dictionary for review
        print("TweepError for id:  " + str(tweet_id))
        fails_dict[tweet_id] = error
        pass
    # To save space, only print out loop/tweet id for every 100 tweet
    if count % 100 == 0:
        print("loop number " + str(count))

# display list of tweets with errors
print("Number of TweepErrors:  {}\n".format(len(fails_dict)))
for tweet_id in fails_dict:
    print(tweet_id, fails_dict[tweet_id])
```

```python
# stack over flow

# creat JSON file
tweet_json_file = 'tweet_json.txt'

# save the JSON data in the list to the output file
with open(tweet_json_file, 'w') as outfile:
    for tweet_json in tweet_json_list:
        json.dump(tweet_json, outfile)
        outfile.write('\n')
#get jason data form text file then convert to df 

tweet_data = []

with open(tweet_json_file, 'r') as json_file:
    # read the first line to start the loop
    line = json_file.readline()
    while line:
        data = json.loads(line)

        # extract variables from the JSON data
        data_id = data['id']
        data_retweet_count = data['retweet_count']
        data_favorite_count = data['favorite_count']
        
        # create a dictionary with the JSON data, then add to a list
        json_data = {'tweet_id': data_id, 
                     'retweet_count': data_retweet_count, 
                     'favorite_count': data_favorite_count
                    }
        tweet_data.append(json_data)

        # read the next line of JSON data
        line = json_file.readline()
        # for the while loop

# converting tweet data dictionary list to a DataFrame
df_tweet_data = pd.DataFrame(tweet_data, 
                                   columns = ['tweet_id',
                                              'retweet_count',
                                              'favorite_count'])

df_tweet_data.head()
```

## Define 3.1 (Tidiness issues)
- Merge multiple dog stages


## Code

```python
# 4 unnecessary columns  [doggo, floofer, pupper, puppo] and be one column called [stage]

Copy_twitter_archive_df.doggo.replace('None', '', inplace=True)
Copy_twitter_archive_df.doggo.replace(np.NaN, '', inplace=True)
Copy_twitter_archive_df.floofer.replace('None', '', inplace=True)
Copy_twitter_archive_df.floofer.replace(np.NaN, '', inplace=True)
Copy_twitter_archive_df.pupper.replace('None', '', inplace=True)
Copy_twitter_archive_df.pupper.replace(np.NaN, '', inplace=True)
Copy_twitter_archive_df.puppo.replace('None', '', inplace=True)
Copy_twitter_archive_df.puppo.replace(np.NaN, '', inplace=True)
Copy_twitter_archive_df.groupby(["doggo", "floofer", "pupper", "puppo"]).size().reset_index().rename(columns={0: "count"})
```
